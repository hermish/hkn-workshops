## 3 Plotting

Matplotlib is a python library that help us to plot data: the most basic plots are line, scatter and histogram plots.

- Line plots usually used when the x-axis is time
- Scatterplots are better when we want to check if there is correlation between two variables
- Histograms visualize the distribution of numerical data


Matplotlib allows us to customize the colors, labels, thickness of lines, title, opacity, grid, figure size... Basically, we can make any graph conceivable, though some are easier than others.

```python
# LINE PLOTS
data['Speed'].plot(
    kind='line', # Type of plot
    color='blue', # Line color
    label='Speed', # Legend name
    linewidth=1,
    alpha = 0.5, # Transparency
    grid = True
)
data['Defense'].plot(
    color='green',
    label='Defense',
    linewidth=1,
    alpha = 0.5,
    grid = True
)

plt.legend(loc='upper right') # legend locations
plt.xlabel('Pokemon Number') # x-axis label
plt.ylabel('Stat') # y-axis label
plt.title('Speed and Defense across Pokemon') # Title
plt.show()
```

```python
# SCATTER PLOT
data.plot(
    kind='scatter', # Type of plot
    x='Attack', # x-axis column
    y='Defense', # y-axis column
    alpha = 0.5, # Transparency
    color = 'red'
)
plt.xlabel('Attack') # x-axis label
plt.ylabel('Defense') # y-axis label
plt.title('Attack-Defense Relationship') # Title
plt.show()
```

```python
# HISTOGRAM
data['Speed'].plot( # Choosing what to plot
    kind = 'hist', # Type of plot
    bins = 50,
    figsize = (15, 15) # Size of graph
)
plt.show()
```

## Data Analysis

Now we can analyze of the data on the pokemonthe easiest way to get an idea of how how data looks is to plot some of it and get a few facts about it. Luckily for us, it is really easy to get the following from our table:

- count: number of entries
- mean: the average numberr, over all pokemon
- std: standard deviation, measures how spread out the data is
- min: minimum entry
- 25%: first quantile (25% are below)
- 50%: median or second quantile (50% are below)
- 75%: third quantile (75% are below)
- max: maximum entry

```python
data.describe()
```

From this overview, we can easily see the minimimum and maximum for each of these statistics. To visualize this, we often use a boxplot, which allows us to easily picture the average, spread and outliers. The black bars the top and bottom show the maximum and minimum data points. The red line in the middle is at the mean and the box shows the distance from the 25th to the 75th percentile.

```python
data.boxplot(
    column='Attack'
)
plt.show()
```

Notice that some of our pokemon are legendary while others are not. We might guess that the attack of legendary pokemon are greater than non-legendary pokemon. We could test our hypothesis by creating a boxplot for the attacks of each group. We can easily do this using the `by` keyword.

```python
data.boxplot(
    column='Attack', # Column plotting
    by='Legendary' # Group
)
plt.show()
```

# 5 Challenge

Finally, there's a lot more interesting figures we can make using python. For example, when we plotted attack and defense they seemed to be linked or *correlated*.

In general, we can test how strongly two variables are correlated they are using the "Pearson product-moment correlation coefficient." This number will always be between +1 and -1. More extreme numbers (near $\pm1$) mean the correlation is strong, while numbers closer to zero indicate that the variables are less strongly linked.

We can visualize how all the variables are correlated by plotting these values in a heat map. The more strongly correlated values are more strongly colored.

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.heatmap(
    data.corr(),
    annot=True,
    linewidths=.5,
    fmt= '.1f',
    ax=ax
)
plt.show()
```

# 6 Applications to Statistical Learning

Now we turn to using our skills on "real-world" data with the goal of making accurate predictions based on data. This is the essense of machine learning, which is described as the "branch of artificial intelligence based on the idea that systems can learn from data, identify patterns and make decisions with minimal human intervention."

Here we are looking at orthopedic patient data with such features as "pelvic radius" and "lumbar lordosis angle" and then making a prediction as to whether a patient's condition is "normal" or "abnormal".

```python
data = pd.read_csv('data/column_2C_weka.csv')
print(data.shape)
data.head(10)
```

# 6.1 Scatter Plots

Let's make some scatter plots comparing features on the x and y axes and plotting a red star for "abnormal" and a green star for "normal". 

```python
color_list = ['red' if i=='Abnormal' else 'green' for i in data.loc[:,'class']]
pd.plotting.scatter_matrix(data.loc[:, data.columns != 'class'],
                                       c=color_list,
                                       figsize= [15,15],
                                       diagonal='hist',
                                       alpha=0.5,
                                       s = 200,
                                       marker = '*',
                                       edgecolor= "black")
plt.show()
```

A question to ask here is: which features seem to be best for classifying normal vs. abnormal? In which plots do we see the best separation of these two labels? 

# 6.2 Making a KNN model

Now, we will make a model for predicting the label of a patient given his/her feature data (notice that there are 6 features such that a patient's feature data can be written as a vector, $\vec{v} \in \mathbb{R}^6$). 

We will use a k-nearest neighbors model. To predict the label of a test data point, this model essentially looks at the training data and picks the k nearest neighbors (i.e. the k training data points with smallest euclidean distance from the test data point; recall that a data point can be represented as a "feature vector"). The model then just assigns the test data point the label that is most common among its k-nearest neighboring training data points.

```python
# We import this KNN model from sklearn
from sklearn.neighbors import KNeighborsClassifier
x,y = data.loc[:,data.columns != 'class'], data.loc[:,'class']
```

Now we split the data into two parts. Here 70% of the data will remaining training data (x_train, y_train) and the other 30% will be reserved for testing (x_test, y_test). Splitting the data into testing and training sets is helpful because it allows us to see how well our model is generalizing and helps us prevent overfitting.

```python
# train test split
from sklearn.model_selection import train_test_split
# randomly take 30% of the data to be test data (we do not train using test data)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
knn = KNeighborsClassifier(n_neighbors = 3)
x,y = data.loc[:,data.columns != 'class'], data.loc[:,'class']
knn.fit(x_train,y_train)
# we predict on our test data to find the accuracy
print('KNN (K=3) accuracy is: ',knn.score(x_test,y_test)) # accuracy
```

Let's look at how changing the number of neigbors affects the accuracy of our prediction. 

```python
# hyperparameter tuning
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors = k)
    x,y = data.loc[:,data.columns != 'class'], data.loc[:,'class']
    knn.fit(x_train,y_train)
    print('KNN (K=' + str(k) + ') accuracy is: ',knn.score(x_test,y_test)) # accuracy
```

Now, let's do what we said we weren't going to do and test with our training data instead of our testing data. 

```python
# why we don't test on training data
for k in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors = k)
    x,y = data.loc[:,data.columns != 'class'], data.loc[:,'class']
    knn.fit(x_train,y_train)
    # notice we are predicting on our X_TRAIN data here
    print('KNN (K=' + str(k) + ') accuracy is: ',knn.score(x_train,y_train)) # accuracy
```
